### Copy an Image

To copy an image, use the copy() method!

Why do this? Well without creating a copy of the img, and simply assigning the img to another variable still point to the same location in memory, so copying creates a copy eg.

```python
new_img_var = img
```

```python
id(img) == id(new_img_var)
```

```python
print(id(img))
```

```python
new_img_copy = img.copy()
```

```python
print(f"Img ID: {id(img)}\nNew Img Copy ID: {id(new_img_copy)}")
```

### Image Rotation, Flipping and Transposition

Flipping images is essentially reordering the indicies of the pixels. So...

```python
# change the image to an array
img_arr = np.array(img)
```

```python
# sanity check

img_arr.shape
```

```python
width, height, c = img_arr.shape
print('width, height, c', width, height, c)
```

```python
# create an array of same dimnesions all zeros
array_flip = np.zeros((width, height, c), dtype=np.uint8)
```

```python
# assign first row of pixels to last row of new frame
for i,row in enumerate(img_arr):
    array_flip[width - 1 - i, :, :] = row
```

```python
plt.figure(figsize=(10,10))
plt.imshow(array_flip)
plt.show()
```
