```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
```

```python
# sauce -> https://archive.ics.uci.edu/dataset/165/concrete+compressive+streng
df = pd.read_csv("../Datasets/concrete_data/Concrete_Data.csv", encoding="utf-8")
```

```python
# is clean?
# df.isnull().sum()
concrete_cols = df.columns
print(concrete_cols)
```

```python
# split data into predictors and target
concrete_cols = df.columns

predictors = df[concrete_cols[concrete_cols != "Concrete compressive strength(MPa, megapascals) "]]
target = df["Concrete compressive strength(MPa, megapascals) "]
```

```python
# sanity check
predictors.head(1)
```

```python
target.head(1)
```

```python
# normlize data by subtracting the mean and dividing by stdDev
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
# save number of predictors to ncols as num required for building network
n_cols = predictors_norm.shape[1] # number of predictors
```

```python
# define function for regression model 
def reg_model():
    # create model
    model = Sequential()
    model.add(Dense(50,activation="relu", input_shape=(n_cols,)))
    model.add(Dense(50,activation="relu"))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer="adam", loss="mean_squared_error")
    return model
```

```python
# build model
model = reg_model()
```

```python
"""
Next, we will train and test the model at the same time using the fit method. 
We will leave out 30% of the data for validation and we will train the model for 100 epochs.
"""
model.fit(predictors_norm, target, validation_split=0.3, epochs=100, verbose=2)
```
