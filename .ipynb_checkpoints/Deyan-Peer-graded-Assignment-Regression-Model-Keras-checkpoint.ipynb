```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
concrete_data = pd.read_csv('concrete_data.csv')
concrete_data.head()
```

```python
concrete_data.isnull().sum()
```

We have no empty values in our data.

### A. Build a baseline model (5 marks)

Use the Keras library to build a neural network with the following: 

- One hidden layer of 10 nodes, and a ReLU activation function 
- Use the adam optimizer and the mean squared error as the loss function. 

#### A1. Randomly split the data into a training and test sets by holding 30% of the data for testing. You can use the train_test_split helper function from Scikit-learn. 

```python
# First we separate the data into predictors and target
concrete_data_columns = concrete_data.columns

predictors = concrete_data.drop(columns=['Strength'])
target = concrete_data['Strength']
n_cols = predictors.shape[1]
```

```python
predictors.head()
```

```python
target.head()
```

```python
# We split the data using the train_test_split function from sklearn

X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=42)
```

#### A2. Train the model on the training data using 50 epochs. 

```python
# We build the model

def regression_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# We train the model

model = regression_model()
model.fit(X_train, y_train, epochs=50, verbose=1)
```

#### A3. Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength. You can use the mean_squared_error function from Scikit-learn. 

```python
# We evaluate and make a prediction
model.evaluate(X_test, y_test)
y_pred = model.predict(X_test)
```

```python
# We calculate the MSE with the function from sklearn

mse = mean_squared_error(y_test, y_pred)

print("Mean Squared Error:", mse)
```

#### A4. Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors. 

```python
# We define a function to make the process faster

def process(process_predictors, process_target, model, epochs):
    # 1. Randomly split the data
    X_train, X_test, y_train, y_test = train_test_split(process_predictors, process_target, test_size=0.3, random_state=42)
    
    # 2. Train the model
    model.fit(X_train, y_train, epochs=epochs, verbose=1)
    
    # 3. Evaluate the model
    model.evaluate(X_test, y_test)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)

    return mse
```

```python
mse_results_A = []
model = regression_model()
epochs = 50

for _ in range(50):
    mse_results_A.append(process(predictors, target, model, epochs))
```

#### A5. Report the mean and the standard deviation of the mean squared errors.

```python
mse_mean_A = np.mean(mse_results_A)
mse_std_A = np.std(mse_results_A)

print('Mean of Mean Squared Errors', mse_mean_A)
print('Standart Deviation of Mean Squared Errors', mse_std_A)
```

```python
# We create an array to store all values so we can plot them later

mse_means = [mse_mean_A]
mse_stds = [mse_std_A]
```

### B. Normalize the data (5 marks) 

Repeat Part A but use a normalized version of the data. Recall that one way to normalize the data is by subtracting the mean from the individual predictors and dividing by the standard deviation. 

#### How does the mean of the mean squared errors compare to that from Step A?
Both mean and standart deviation have decreased, going from 69 and 60 in A to 55 and 53 in B.

```python
# We normalize the data

mean_values = np.mean(predictors, axis=0)
std_deviation_values = np.std(predictors, axis=0)

predictors_normalized = (predictors - mean_values) / std_deviation_values
```

```python
predictors_normalized.head()
```

```python
# Run the model with the normalized data

mse_results_B = []
model = regression_model()
epochs = 50

for _ in range(50):
    mse_results_B.append(process(predictors_normalized, target, model, epochs))
```

```python
mse_mean_B = np.mean(mse_results_B)
mse_means.append(mse_mean_B)
mse_std_B = np.std(mse_results_B)
mse_stds.append(mse_std_B)

print('Mean of Mean Squared Errors with normazed data', mse_mean_B)
print('Standart Deviation of Mean Squared Errors with normalized data', mse_std_B)
```

C. Increate the number of epochs (5 marks) 

Repeat Part B but use 100 epochs this time for training. 

#### How does the mean of the mean squared errors compare to that from Step B?
Again, both values have decreased, going from 55 and 53 in B to 41 and 19 in C.

```python
mse_results_C = []
model = regression_model()
epochs = 100

for _ in range(50):
    mse_results_C.append(process(predictors_normalized, target, model, epochs))
```

```python
mse_mean_C = np.mean(mse_results_C)
mse_means.append(mse_mean_C)
mse_std_C = np.std(mse_results_C)
mse_stds.append(mse_std_C)

print('Mean of Mean Squared Errors with normazed data and 100 epochs', mse_mean_C)
print('Standart Deviation of Mean Squared Errors with normalized data and 100 epochs', mse_std_C)
```

D. Increase the number of hidden layers (5 marks) 

Repeat part B but use a neural network with the following instead: 
- Three hidden layers, each of 10 nodes and ReLU activation function. 

#### How does the mean of the mean squared errors compare to that from Step B?
We obtain better results in comparison with step B, going from 55 and 53 in B to 42 and 20 in D.

```python
# We create a new regression model

n_cols = X_train.shape[1]

def regression_model_D():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
mse_results_D = []
model = regression_model_D()
epochs = 50 # We use 50 epochs since those were indicated in part B

for _ in range(50):
    mse_results_D.append(process(predictors_normalized, target, model, epochs))
```

```python
mse_mean_D = np.mean(mse_results_D)
mse_means.append(mse_mean_D)
mse_std_D = np.std(mse_results_D)
mse_stds.append(mse_std_D)

print('Mean of Mean Squared Errors with normazed data, 50 epochs and three hidden layers', mse_mean_D)
print('Standart Deviation of Mean Squared Errors with normalized data, 50 epochs and three hidden layers', mse_std_D)
```

We finish off by plotting all the results to have a more visual understanding of the results.

```python
plt.bar(labels, mse_stds)

plt.xlabel('Processes')
plt.ylabel('Standart Deviation of Mean Squared Error (MSE)')
plt.title('Comparison of Standart Deviation of MSE between Processes A to D')

# Show the plot
plt.show()
```
